```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
# Set the parameters for the simulation
L = 1  # size of the box
x = np.linspace(0, L, 20)
y = np.linspace(0, L, 20)
z = np.linspace(0, L, 20)
nx, ny, nz = 1, 2, 3  # quantum numbers for each dimension
A = (8/L)**0.5

def square_wave_function(x, y, z, nx, ny, nz, L):
    """
    The wave function for a particle in a 3D box
    """
    psi = A*np.sin(nx * np.pi * x / L) * np.sin(ny * np.pi * y / L) * np.sin(nz * np.pi * z / L)
    psi_2 = psi**2
    return psi_2

# Set the parameters for the simulation
L = 1  # size of the box
Xs = []
for i in range(0,3):
   P = int(input("Enter the Quantam numbers:"))
   Xs.append(P)
print("For quantam numbers:")
print(*Xs, sep = ",") 


nx, ny, nz = [Xs[i] for i in (0,1,2)]


x, y, z = np.meshgrid(x,y,z)
# Calculate the wave function for the particle in the box
w = square_wave_function(x, y, z, nx, ny, nz, L)
%matplotlib notebook
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

img = ax.scatter(x, y, z, s=w*2, c=w*3 , cmap='YlGn')
fig.colorbar(img)
plt.show()
```
